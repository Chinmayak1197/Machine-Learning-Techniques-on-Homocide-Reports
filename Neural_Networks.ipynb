# Reading & Pre-processing

```python
data = pd.read_csv('/content/drive/MyDrive/ML DS539/CoLab/training.csv')
test_data = pd.read_csv('/content/drive/MyDrive/ML DS539/CoLab/testing.csv')
```

```python
data = data.drop(['state', 'Next_Year_Top_10'], axis=1)
test_data = test_data.drop(['state', 'Next_Year_Top_10'], axis=1)
```

```python
X = data.drop("Next_Year_Crime_Rate", axis=1)
Y = data["Next_Year_Crime_Rate"]
```

```python
X_test = test_data.drop("Next_Year_Crime_Rate", axis=1)
Y_test = test_data["Next_Year_Crime_Rate"]
```

```python
X["year"] = [(1 + (i-1980)) for i in X["year"]]
X_test["year"] = [(1 + (i-1980)) for i in X_test["year"]]

#This allows for the scaling of the years to be the same (i.e. 1980 is the same distance away from 1981 as 1981 is from 1982) and the timeline/linearity to stay the same as well (i.e. 1980 < 1981)
```

# Artifical Neural Networks



```python
ann = tf.keras.models.Sequential()   #creating a sequential layer
```

Adding Hidden Layers

```python
ann.add(tf.keras.layers.Dense(units=10, input_dim=X.shape[1], activation="relu"))   

#reLu: piecewise linear function that will output the input directly if it is positive, otherwise, it will output zero
```

```python
ann.add(tf.keras.layers.Dense(units=10, activation="relu"))
```

```python
#ann.add(tf.keras.layers.Dense(units=5, activation="relu"))
```

```python
#ann.add(tf.keras.layers.Dense(units=5, activation="relu"))
```

```python
ann.add(tf.keras.layers.Dense(units=1))     #output layer consisting of one neuron
```

```python
ann.summary()
#sequential model summary
```

Compiling and fitting the model on the training set

```python
ann.compile(optimizer='adam', loss='mse')

#adam: combines the best properties of the AdaGrad and RMSProp algorithms to provide an optimization algorithm that can handle sparse gradients on noisy problems
```

```python
ann.fit(X, Y, epochs=500)
```

Predicting the values for test set and calculating the mean error between the actual and predicted values

```python
y_pred = ann.predict(X_test)
```

```python
from sklearn import metrics
print ("mean_absolute_error :",metrics.mean_absolute_error(Y_test, y_pred))
#print("\n")
print ("mean_squared_error : ",metrics.mean_squared_error(Y_test, y_pred))
#print("\n")
print ("root_mean_squared_error : ",np.sqrt(metrics.mean_squared_error(Y_test, y_pred)))
```
